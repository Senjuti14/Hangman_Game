```python
#Hangman game!

#import needed modules
import random

#hangman visuals
hangman = [
"""
+---+
    |
    |
    |
    ===""", """
+---+
  | |
  O |
    |
    ===""", """
+---+
  | |
  O |
 /| |
    ===""", """
+---+
  | |
  O |
 /|\|
    ===""", """
+---+
  | |
  O |
 /|\|
  |  ===""", """
 / 
+---+
  | |
  O |
 /|\|
  |  ===""", """
 / \
"""
]

#create a word list
word_list=["cat", "dog", "monkey", "glass", "baby", "animal", "February", "Python", "casual", "Empty", "Outstanding"]

#pick the word for guessing randomly
word= random.choice(word_list).lower()

#create empty list for correct and incorrect guesses
correct= []
incorrect= []
user_tries= 6 
count = -1

while user_tries>0:
    result=''
    for n in word:
        if n in correct:
            result+=n
        else:
            result += '_ '
    if result== word:
        break
        
    print("The word to guess is: ", result)
    print(user_tries, "chances are left!")
    
    user_guess= input().lower()

    if user_guess in word:
        print("You guessed a correct letter!")
        correct.append(user_guess)
    else:
        print("Wrong guess!")
        count= count+1
        user_tries= user_tries - 1
        incorrect.append(user_guess)
        print(hangman[count])
        
if user_tries>0:
    print("You guessed the word correctly, Congratulations!")
else:
    print("Sorry! No chances left! You could not guess the word!")
        
```
